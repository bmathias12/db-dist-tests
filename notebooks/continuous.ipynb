```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns

COLORS = ['#004050', '#BA0C2F']
```

# Generate Random Series

```python
mu1, sigma1 = 0, 1
mu2, sigma2 = 0.1, 1
N = 1000

s1 = np.random.normal(mu1, sigma1, N)
s2 = np.random.normal(mu2, sigma2, N)
```

```python
_, ax = plt.subplots(figsize=(10,6))

mean_1 = s1.mean()
mean_2 = s2.mean()

sns.histplot(s1, kde=True, ax=ax, color=COLORS[0], alpha=0.3, label='Old::mean: {:0.2f}'.format(mean_1), common_norm=False)
sns.histplot(s2, kde=True, ax=ax, color=COLORS[1], alpha=0.3, label='New::mean: {:0.2f}'.format(mean_2), common_norm=False)
plt.legend()
plt.show()
```

# Tests

## T-Test
Tests that the mean of `s2` is different from the mean of `s1`

```python
stats.ttest_ind(s1, s2)
    
```

## Two-Sample Kolmogorov-Smirnov (KS) Test

The two-sample Kolmogorov-Smirnov test is used to test whether two samples come from the same distribution.

```python
stats.ks_2samp(s1, s2)
```

```python
def plot_ks(s1, s2):
    _, ax = plt.subplots(figsize=(10,6))
    sns.ecdfplot(s1, color=COLORS[0], ax=ax)
    sns.ecdfplot(s2, color=COLORS[1], ax=ax)
    plt.title('Empirical CDF of S1 and S2')

plot_ks(s1, s2)
```
